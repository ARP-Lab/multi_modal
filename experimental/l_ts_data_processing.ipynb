```python
import pandas as pd
import numpy as np
from glob import glob
import pickle
from collections import Counter
```

```python
annot_file_lst = glob('./KEMDy20/annotation/*.csv')
annot_file_lst = sorted(annot_file_lst)
annot_file_lst[:3]
```

```python
# session 1만 해보기
# for i in annot_file_lst:
#     file_lst = pd.read_csv(i, skiprows = 1)[' .1']
import re

for i in range(len(annot_file_lst)):
    # print(annot_file_lst[i])
    if i == 0:
        annot_data = pd.read_csv(annot_file_lst[i], skiprows = 1)
    else:
        annot_data = pd.concat([annot_data, pd.read_csv(annot_file_lst[i], skiprows = 1)], axis= 0)
annot_data.rename(columns = {' .1': 'segment_id'}, inplace= True)
annot_data
```

```python
with open('./model/data/paradeigma_ts_data_preprocessing.pkl', 'rb') as f:
    ts_dataset_by_session = pickle.load(f)
ts_dataset_by_session[1]
```

```python
for i in range(1,40+1):
    if i == 1:
        ts_dataset_flatten = ts_dataset_by_session[i].loc[:,['segment_id','eda','temp']]
    else:
        ts_dataset_flatten = pd.concat([ts_dataset_flatten, ts_dataset_by_session[i].loc[:,['segment_id','eda','temp']]], axis= 0)
ts_dataset_flatten = ts_dataset_flatten.reset_index()
ts_dataset_flatten
```

```python
import math
def add_padding(pd_series, length = 50):
    if isinstance(pd_series, float):
        if math.isnan(pd_series):
            return np.zeros(10)
    if len(pd_series) < length:
        pd_series = np.concatenate([pd_series, np.zeros(length - len(pd_series))])
        return pd_series
    elif len(pd_series) == length:
        return pd_series
    elif len(pd_series) > length:
        pd_series = pd_series[:length]
        return pd_series
```

```python
padded_eda_data = map(add_padding , ts_dataset_flatten['eda'])
ts_dataset_flatten['eda'] = list(padded_eda_data)
padded_temp_data = map(add_padding , ts_dataset_flatten['temp'])
ts_dataset_flatten['temp'] = list(padded_temp_data)

# check
ts_dataset_flatten['eda'][0], ts_dataset_flatten['temp'][0]
```

```python
ts_dataset = pd.merge(annot_data, ts_dataset_flatten, on='segment_id', how='outer')
len(ts_dataset['temp'][0]),len(ts_dataset['eda'][0]) 
```

```python
# 참고: https://stackoverflow.com/questions/27275236/how-to-select-all-columns-whose-names-start-with-x-in-a-pandas-dataframe
filter_col = [col for col in ts_dataset.columns if col.startswith('Emotion.')]
filter_col
Counter(ts_dataset[filter_col].iloc[10])
```

```python
annot_target = []
for i in range(len(ts_dataset)):
    score_table = {'neutral': 0, 'happy': 0, 'surprise':0, 'disgust': 0, 'angry': 0, 'sad':0, 'fear': 0}
    i_of_tsdata = Counter(ts_dataset[filter_col].iloc[i])
    for emotion in i_of_tsdata.keys():
        score_table[emotion] += i_of_tsdata[emotion]
    annot_target.append(list(score_table.values()))
annot_target[:3]
```

```python
ts_dataset['new_emotion'] = annot_target
ts_dataset
```

```python
with open('./model/data/lou_ts_dataset.pkl', 'wb') as f:
    pickle.dump(ts_dataset, f, pickle.HIGHEST_PROTOCOL)
```
